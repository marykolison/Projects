```python
df.fillna(0, inplace=True)
```

```python
# Replace missing values in numerical columns with the mean
numeric_columns = df.select_dtypes(include=['number'])
df[numeric_columns.columns] = df[numeric_columns.columns].fillna(df[numeric_columns.columns].mean())

# For non-numeric columns, you can replace missing values with a specific value, e.g., 'Unknown' for string columns.
string_columns = df.select_dtypes(include=['object'])
df[string_columns.columns] = df[string_columns.columns].fillna('Unknown')
```

```python
plt.figure(figsize=(11,5))
sns.barplot(data=df)
plt.xticks(rotation=45)
plt.title('Global CO2 Emission tredn from 1990 to 2018')
plt.xlabel('Year')
plt.ylabel('Total CO2 Emission Rate (mt)')
plt.show()
```

```python
# Group the DataFrame by 'Year' and calculate the sum of 'CO2 Emission Rate (mt)' for each year
total_emissions_by_year = df.groupby('1990').sum().reset_index()

# Print or display the total emissions by year
print(total_emissions_by_year)
```

```python
#Calculate the total emissions for the specific year (e.g., 2021)
year = '2021'
total_emissions_for_year = df[df['1990'] == int(year)]['2000'].sum()

# Convert emissions for the specific year to numeric data type
df[year] = pd.to_numeric(df[year], errors='coerce')

#Calculate the proportion of emissions contributed by each country for that year
df['Proportion'] = df[year] / total_emissions_for_year

#Sort the DataFrame by the 'Proportion' column in descending order
sorted_df = df.sort_values(by='Proportion', ascending=False)

# Display the sorted DataFrame
print(sorted_df[['1990', year, 'Proportion']])

```

```python
year = '2021' 
top_5 = sorted_df[['1990', 'Proportion']].head(5)

# Display the top 5 contributors
print(top_5)
```

```python
# Check for missing values
missing_values = top_5['1990'].isna().sum()
print(f"Number of missing values in 'Emissions': {missing_values}")

# Drop rows with missing 'Emissions' values
top_5 = top_5.dropna(subset=['1990'])

# Create the bar plot after handling missing values
plt.figure(figsize=(12, 6))
sns.barplot(data=top_5)
plt.title(f'Top 5 CO2 Emitters in {year}')
plt.xlabel('Country')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
